```python
import numpy as np
a = np.array([3.1,1,-0.5,-3.2,6])
b = np.array([1,3,2.2,5.1,1])

mult = a * b
print(mult)

productopunto = np.dot(a,b) # también a @ b
print(productopunto)
```

```python
A = np.array([[2,-1,-3],
              [4,1.5,-2.5],
              [7.3,-0.9,0.2]])

AT = np.transpose(A) # también AT = A.T
print(AT)
```

```python
o = np.ones(A.shape)
print("ones")
print(o)
# crea una matriz de nxm con todos los elementos iguales a 1

r = np.round(A)
print("round")
print(r)
# a core NumPy method for rounding elements of an array to a specified number
# of decimal places

c = np.ceil(A)
print("ceil")
print(c)
# a core NumPy method for rounding elements of an array up to the nearest integer

f = np.floor(A)
print("floor")
print(f)
# a core NumPy method for rounding elements of an array down to the nearest integer

```
