# REPASO NUMPY

```python
import numpy as np
a = np.array([3.1,1,-0.5,-3.2,6])
b = np.array([1,3,2.2,5.1,1])

mult = a * b

print('Vector a:')
print(a)
print('\nVector b:')
print(b)
print('\nProducto a.b:')
print(mult)
```

```python
A = np.array([[2,-1,-3],
              [4,1.5,-2.5],
              [7.3,-0.9,0.2]])

AT = np.transpose(A) # también AT = A.T

print('\nMatriz A:')
print(A)
print('\nMatriz transpuesta AT:')
print(AT)
```

```python
o = np.ones(A.shape)
print("ones")
print(o)
# crea una matriz de nxm con todos los elementos iguales a 1

r = np.round(A)
print("\nround")
print(r)
# redondea los elementos de la matriz al entero más cercano

c = np.ceil(A)
print("\nceil")
print(c)
# redondea los elementos de la matriz hacia arriba al entero más cercano

f = np.floor(A)
print("\nfloor")
print(f)
# redondea los elementos de la matriz hacia abajo al entero más cercano
```

```python
A = np.array([[2,-1,-3],
              [4,1.5,-2.5],
              [7.3,-0.9,0.2]])
print("\nMatriz A:")
print(A)
# acceder a un elemento específico de la matriz
print("\nElemento en la posición (0,1) de la matriz A:")
print(A[0,1])

# acceder a una fila específica de la matriz
print("\nSegunda fila de la matriz A:")   
print(A[1])

# dimensiones de la matriz, se utiliza el atributo .shape
print("\nDimensiones de la matriz A:") 
print(A.shape)
```

```python
# crear la función función  y[n]=sin⁡(pi*0.12n) 

n = np.arange(0,101) #
y = np.sin(np.pi*0.12*n)
y2 = np.cos(2*np.pi*0.03*n)

s = y + y2
t = y * y2
```

```python
import matplotlib.pyplot as plt

# gráfica de las funciones y[n] y y2[n]
plt.figure(figsize=(10,6))
plt.title('Funciones y[n] y y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.plot(n, y, label='y[n] = sin(pi*0.12n)', color='green')
plt.plot(n, y2, label='y2[n] = cos(2pi*0.03n)', color = 'orange')
plt.legend()

# gráfica de las funciones s[n] y t[n]
plt.figure(figsize=(10,6))
plt.title('Funciones s[n] y t[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.plot(n, s, label='s[n] = y[n] + y2[n]', color='blue')
plt.plot(n, t, label='t[n] = y[n] * y2[n]', color='red')
plt.legend()
```

```python
# escribir una función que reciba un diccionario con las notas de los alumnos de un curso
# y devuelva una serie con la nota mínima, la máxima, media y desviación típica.

def calcular_estadisticas(notas):
    notas_series = pd.Series(notas)
    estadisticas = {
        'min': notas_series.min(),
        'max': notas_series.max(),
        'mean': notas_series.mean(),
        'std': notas_series.std()
    }
    return pd.Series(estadisticas)
```
